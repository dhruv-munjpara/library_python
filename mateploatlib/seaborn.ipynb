```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# load  built-in dataset
df=sns.load_dataset("tips")

df.head()
```

```python
df.info()
```

```python
sns.histplot(data=df,x="total_bill",bins=20,kde=True)
plt.show()
```

```python
sns.histplot(data=df,x="total_bill",bins=20)
plt.show()
```

```python
sns.kdeplot(data=df,x="total_bill")
plt.show()
```

```python
sns.distplot(df["total_bill"])
```

```python
# blox ploat
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.boxplot(x="day",y="tip",data=df)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
# pair ploat
sns.pairplot(df)
plt.show()
```

```python
corr=df.corr(numeric_only=True)

sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.relplot(data=df,x="total_bill",y="tip",kind="line")
plt.show()
```

```python
sns.catplot(data=df,x="day",y="total_bill",kind="bar")
plt.show()
```

```python
sns.displot(data=df,x="total_bill",kind="hist")
plt.show()
```

```python
sns.jointplot(data=df,x="total_bill",y="tip")
plt.show()
```

```python
sns.countplot(data=df,x="day")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.histplot(data=df,x="total_bill")
plt.show()
```
