```python
import matplotlib.pyplot as plt
# subplots
x1=[1,2,3,4]
y1=[10,20,15,25]

plt.plot(x1,y1)
plt.scatter(x1,y1)
plt.bar(x1,y1)

plt.title("using figure and axes")
plt.xlabel("x values")
plt.ylabel("y values")

plt.show()
```

```python
x=[1,2,3,4]
y1=[10,20,30,40]
y2=[15,25,35,45]

plt.plot(x,y1,label="line1")
plt.plot(x,y2,label="line2")

plt.title("multiple line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.legend()

plt.show()
```
