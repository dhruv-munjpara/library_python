```python
# data cleaning

# chek missing values
df.isnull().sum()
```

```python
df["Engine Fuel Type"]=df["Engine Fuel Type"].fillna(df["Engine Fuel Type"].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df["Engine HP"]=df["Engine HP"].fillna(df["Engine HP"].median())
df
```

```python
df.isnull().sum()
```

```python
df["Engine Cylinders"]=df["Engine Cylinders"].fillna(df["Engine Cylinders"].median())
df
```

```python
df.isnull().sum()
```

```python
df["Number of Doors"]=df["Number of Doors"].fillna(df["Number of Doors"].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df["Market Category"]=df["Market Category"].fillna("Unknow")
df
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.columns=df.columns.str.lower().str.replace(" ","_")
df
```

```python
df.info()
```

```python
df["engine_cylinders"]=df["engine_cylinders"].astype(int)
df["number_of_doors"]=df["number_of_doors"].astype(int)
```

```python
df.info()
```

```python
# handle outliers
Q1=df["msrp"].quantile(0.25)
Q3=df["msrp"].quantile(0.75)

IQR=Q3-Q1

df=df[(df["msrp"]>=Q1-1.5*IQR)&(df["msrp"]<=Q3+1.5*IQR)]
df
```
