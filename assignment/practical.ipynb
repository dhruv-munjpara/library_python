```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df=pd.read_csv("data.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
# data cleaning

# chek missing values
df.isnull().sum()
```

```python
df["Engine Fuel Type"]=df["Engine Fuel Type"].fillna(df["Engine Fuel Type"].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df["Engine HP"]=df["Engine HP"].fillna(df["Engine HP"].median())
df
```

```python
df.isnull().sum()
```

```python
df["Engine Cylinders"]=df["Engine Cylinders"].fillna(df["Engine Cylinders"].median())
df
```

```python
df.isnull().sum()
```

```python
df["Number of Doors"]=df["Number of Doors"].fillna(df["Number of Doors"].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df["Market Category"]=df["Market Category"].fillna("Unknow")
df
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.columns=df.columns.str.lower().str.replace(" ","_")
df
```

```python
df.info()
```

```python
df["engine_cylinders"]=df["engine_cylinders"].astype(int)
df["number_of_doors"]=df["number_of_doors"].astype(int)
```

```python
df.info()
```

```python
# handle outliers
Q1=df["msrp"].quantile(0.25)
Q3=df["msrp"].quantile(0.75)

IQR=Q3-Q1

df=df[(df["msrp"]>=Q1-1.5*IQR)&(df["msrp"]<=Q3+1.5*IQR)]
df
```

```python
df.describe()
```

```python
print("Mean MSRP:", np.mean(df["msrp"]))
print("Max MSRP:", np.max(df["msrp"]))
print("Min MSRP:", np.min(df["msrp"]))
```

```python
print(df[["engine_hp","msrp"]].corr())
```

```python
# for visualization
plt.Figure(figsize=(8,5))
plt.hist(df["msrp"],bins=30)
plt.title("distribution of price")
plt.xlabel("price")
plt.ylabel("freq")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x="engine_hp", y="msrp", data=df, alpha=0.6)  #alpha =reduce overlap

plt.title("Engine HP vs Price", fontsize=15)
plt.xlabel("Engine Horsepower")
plt.ylabel("Price")
plt.grid(True)   #gride=cleare reading

plt.show()
```

```python
sns.boxplot(x=df["msrp"])
plt.title("Boxplot of Price")
plt.show()
```

```python
sns.countplot(x="number_of_doors", data=df)
plt.title("Doors Count")
plt.show()
```

```python
sns.pairplot(df[["engine_hp", "engine_cylinders", "msrp"]])
plt.show()
```
