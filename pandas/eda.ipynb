```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of rows
n = 500

# Generate dataset
data = {
    "Customer_ID": np.random.randint(1, 101, n),  # 100 customers
    "Region": np.random.choice(
        ["North","South","East","West","Central"], 
        n, 
        p=[0.25,0.2,0.2,0.2,0.15]
    ),
    "Product": np.random.choice(
        ["Laptop","Mouse","Tablet","Phone","Keyboard"], 
        n
    ),
    "Category": np.random.choice(
        ["Electronics","electronics","ELECTRONICS", None], 
        n,
        p=[0.4,0.3,0.2,0.1]
    ),
    "Quantity": np.random.randint(1, 10, n),
    "Price": np.random.randint(500, 80000, n).astype(float),
    "Order_Date": pd.date_range(start="2024-01-01", periods=n, freq="D")
}

df = pd.DataFrame(data)

# Introduce missing values in Price
missing_indices = np.random.choice(df.index, size=30, replace=False)
df.loc[missing_indices, "Price"] = np.nan

# Shuffle dataset
df = df.sample(frac=1).reset_index(drop=True)

# Display dataset
print("head")
df.head()
```

```python
# part 1=numpy practice
# create a numpy array of quantity columm
qty_array=df["Quantity"].to_numpy()
qty_array
```

```python
# 2 Perform:

# * Total quantity sold  
# * Average quantity  
# * Maximum quantity  
# * Minimum quantity
total_qty=np.sum(qty_array)
avg_qty=np.mean(qty_array)
max_qty=np.max(qty_array)
min_qty=np.min(qty_array)
total_qty,avg_qty,max_qty,min_qty
```

```python
# part-2 data understanding
# 1 load the dataframe
df = pd.DataFrame(data)
df
```

```python
# 3.check the datatype
df.dtypes
```

```python
# check the missing values
df.isnull().sum()
```

```python
# 3-data cleaning
# 1.fill missing values
df["Category"]=df["Category"].fillna("Unknow")
df
```

```python
df.isnull().sum()
```

```python
df["Price"]=df["Price"].fillna(df["Price"].mean()).round(2)
df
```

```python
df.isnull().sum()
```

```python
# 2.convert category to lowercase
df["Category"] = df["Category"].str.lower()
df
```

```python
# 3.convet order_date to datetime
df["Order_Date"]=pd.to_datetime(df["Order_Date"])
df
```

```python
# 4.extract year and month
df["Year"]=df["Order_Date"].dt.year
df
```

```python
df["Month"]=df["Order_Date"].dt.month
df
```

```python
# 4.feature engineering
df["Revenue"]=df["Quantity"]*df["Price"].round(2)
df
```

```python
# part-5=groupby analysis
# 1.total revenue by region
df.groupby("Region")["Revenue"].sum()
```

```python
# 2.total revenue by product
df.groupby("Product")["Revenue"].sum()
```

```python
# 3.avg revenue by category
df.groupby("Category")["Revenue"].mean()
```

```python
# 4.multi-level groping
df.groupby(["Region","Category"])["Revenue"].sum()
```

```python
# part-6=pivot table:
pivot_table=pd.pivot_table(df,values="Revenue",index="Region",columns="Product",aggfunc="sum",fill_value=0)
pivot_table
```

```python
# part-7=sorting and insights
# 1.region with highest revenue
re_wi_re=df.groupby("Region")["Revenue"].sum().sort_values(ascending=False).head(1)
re_wi_re
```

```python
# 2.product with highest revenue
pro_wi_re=df.groupby("Product")["Revenue"].sum().sort_values(ascending=False).head(1)
pro_wi_re
```

```python
# 3.sorrt regions by revenue
df.groupby("Region")["Revenue"].sum().sort_values(ascending=False)
```

```python

# part-8=visualization
import matplotlib.pyplot as plt

monthly_revenue=df.groupby("Month")["Revenue"].sum()
plt.Figure()
plt.plot(monthly_revenue.index,monthly_revenue.values,marker='o')
plt.title("monthly revenue trend")
plt.xlabel("Month")
plt.ylabel("Total revenue")
plt.show()
```
