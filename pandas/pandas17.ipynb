```python
import pandas as pd
data={
    "region":["north","south","north","south","east"],
    "category":["a","a","b","b","a"],
    "revenue":[100,200,150,250,300]
}
df=pd.DataFrame(data)
df
```

```python
df.groupby("region").sum()
```

parameter of groupby
df.groupby(by,axis=0,level=none,as_index=True,group_keys=True,observed=False,dropna=True)
1.by=column name(s) used for grouping
ex:df.groupby("rigion").

```python
df.groupby("region",as_index=False).sum()
```

```python
df.groupby(["region","category"]).sum()
```

>aggregation function
1.sum()
2.mean()
3.count()
4.max()
5.min()
6.std()

```python
df.groupby("region")["revenue"].mean()
```

```python
df.groupby("region").agg({
    "revenue":"sum"
})
```

```python
df.groupby("region").agg({
    "revenue":["sum","mean","std","min","max"]
})
```

```python
df.groupby("region").agg(
    total_revenue=("revenue","sum"),
    avg_revenue=("revenue","mean")
)
```

```python
df.groupby(["region","category"])["revenue"].sum()
```

```python
df.groupby(["region","category"])["revenue"].sum().reset_index()
```
