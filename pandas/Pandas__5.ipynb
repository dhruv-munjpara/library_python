# SESSION 10 — Merging & Joining DataFrames

## Why Merging is Important

In real-world scenarios, data is often stored in **multiple tables/files**.

Examples:

Customer Table:

| Customer_ID | Name |
|-------------|------|
| 1 | Aman |
| 2 | Priya |

Sales Table:

| Customer_ID | Amount |
|-------------|--------|
| 1 | 500 |
| 2 | 700 |

To analyze data properly, we need to **combine these tables**.

---

# Solution

Pandas provides:

merge()

---

# Key Concept

Merge is similar to **SQL JOIN**

# merge() Function

## Syntax

pd.merge(left_df, right_df, how="type", on="column")

---

## Parameters

left_df → first DataFrame  
right_df → second DataFrame  

how → type of join:

• "inner" → common values only  
• "left" → all values from left  
• "right" → all values from right  
• "outer" → all values from both  

on → column name used for joining  

---

# Use Case

• Combine customer data with sales  
• Combine datasets from multiple sources

```python
import pandas as pd

customers = pd.DataFrame({
    "Customer_ID": [1,2,3],
    "Name": ["Aman","Priya","Rahul"]
})

sales = pd.DataFrame({
    "Customer_ID": [1,2,4],
    "Amount": [500,700,300]
})

customers, sales
```

# 1. Inner Join

Keeps only matching values in both tables.

```python
pd.merge(customers, sales, how="inner", on="Customer_ID")
```

# 2. Left Join

Keeps all rows from left table.

```python
pd.merge(customers, sales, how="left", on="Customer_ID")
```

# 3. Right Join

Keeps all rows from right table.

```python
pd.merge(customers, sales, how="right", on="Customer_ID")
```

# 4. Outer Join

Keeps all rows from both tables.

```python
pd.merge(customers, sales, how="outer", on="Customer_ID")
```

# Merge on Multiple Columns

## Syntax

pd.merge(df1, df2, on=["col1","col2"])

---

## Use Case

When one column is not enough to uniquely identify data.

```python
df1 = pd.DataFrame({
    "ID":[1,1,2],
    "City":["Delhi","Mumbai","Delhi"],
    "Sales":[100,200,300]
})

df2 = pd.DataFrame({
    "ID":[1,2],
    "City":["Delhi","Delhi"],
    "Target":[150,350]
})

pd.merge(df1, df2, on=["ID","City"])
```

# Default Behavior of pd.merge()

When you do:

pd.merge(df1, df2, on=["ID","City"])

and do NOT specify `how`

---

# Default Join Type

The default join is:

INNER JOIN

---

# What is INNER JOIN?

It keeps only those rows where:

• Values match in BOTH DataFrames  
• Based on the given key(s)

---

# Important Rule

If no match exists → row is dropped

---

# Syntax Reminder

pd.merge(df1, df2, how="inner", on=["ID","City"])

is SAME as:

pd.merge(df1, df2, on=["ID","City"])

# Demo — Customer + Sales Merge

Goal:

Combine customer details with their sales data.

```python
merged_df = pd.merge(customers, sales, how="left", on="Customer_ID")

merged_df
```

# SESSION 11 — Concatenation & Data Reshaping

## Why Concatenation is Needed

Sometimes data is split across multiple files:

Example:

January Sales  
February Sales  
March Sales  

We need to **combine them into one dataset**.

---

# Solution

pd.concat()

# pd.concat()

## Syntax

pd.concat([df1, df2], axis=0)

---

## Parameters

axis:

0 → row-wise (default)  
1 → column-wise  

ignore_index:

True → reset index  

---

# Use Cases

• Combine datasets  
• Append rows  
• Merge columns

```python
df1 = pd.DataFrame({"Sales":[100,200]})
df2 = pd.DataFrame({"Sales":[300,400]})

pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

# Column-wise Concatenation

axis = 1

```python
df1 = pd.DataFrame({"A":[1,2]})
df2 = pd.DataFrame({"B":[3,4]})

pd.concat([df1, df2], axis=1)
```

# Pivot

## Purpose

Reshape data from long format to wide format.

---

## Syntax

df.pivot(index, columns, values)

```python
data = {
    "Month":["Feb","Feb","Jan","Jan"],
    "Product":["A","B","A","B"],
    "Sales":[100,200,150,250]
}

df = pd.DataFrame(data)

df.pivot(index="Month", columns="Product", values="Sales")
```

# Pivot Table

## Purpose

Similar to pivot but supports aggregation.

---

## Syntax

df.pivot_table(index, columns, values, aggfunc)

```python
df.pivot_table(index="Month", columns="Product", values="Sales", aggfunc="sum")
```

# Pivot

## What Problem Does Pivot Solve?

Sometimes data is stored in **long format**:

| Month | Product | Sales |
|------|--------|------|
| Jan  | A      | 100  |
| Jan  | B      | 200  |
| Feb  | A      | 150  |
| Feb  | B      | 250  |

This is called **long / tidy format**.

---

## What Do We Want?

We want something like:

| Month | A   | B   |
|------|----|----|
| Jan  | 100 | 200 |
| Feb  | 150 | 250 |

This is called **wide format**.

---

## What Pivot Does

### Pivot rearranges data:

Rows → become index  
Unique column values → become new columns  
Values → fill the table  

---

# Core Idea

pivot = "reshape table"

```python
import pandas as pd

# Step 1: Create long format dataset
data = {
    "Month": ["Jan","Jan","Feb","Feb","Mar","Mar"],
    "Product": ["A","B","A","B","A","B"],
    "Sales": [100,200,150,250,300,400]
}

df = pd.DataFrame(data)

print("Original Data (Long Format):\n")
print(df)
```

# Step-by-Step Pivot Explanation

We want:

• Month → rows  
• Product → columns  
• Sales → values  

---

# Syntax

df.pivot(index="Month", columns="Product", values="Sales")

---

# Parameter Meaning

index → what becomes rows  
columns → what becomes columns  
values → what fills the table

```python
pivot_df = df.pivot(index="Month", columns="Product", values="Sales")

print("\nPivot Output (Wide Format):\n")
print(pivot_df)
```

# Output Understanding

| Month | A   | B   |
|------|----|----|
| Jan  | 100 | 200 |
| Feb  | 150 | 250 |
| Mar  | 300 | 400 |

---

# How It Worked

For each combination:

Month + Product → value is picked from Sales

Example:

Month = Jan, Product = A → 100  
Month = Jan, Product = B → 200  

---

# Important Rule (VERY IMPORTANT)

Pivot works ONLY when:

Each (index, column) pair has ONE value

If duplicates exist → ERROR

```python
# Duplicate example (WILL FAIL)

data = {
    "Month": ["Jan","Jan"],
    "Product": ["A","A"],
    "Sales": [100,150]
}

df = pd.DataFrame(data)

# This will raise error
df.pivot(index="Month", columns="Product", values="Sales")
```

# Why Pivot Fails Here?

Because:

Month = Jan, Product = A has TWO values:

100 and 150

- Pivot does not know which one to choose

---

# Solution

Use pivot_table() instead

```python
df.pivot_table(index="Month", columns="Product", values="Sales", aggfunc="sum")
```

# Pivot vs Pivot Table

| Feature | pivot | pivot_table |
|--------|------|------------|
| Handles duplicates |  No |  Yes |
| Aggregation |  No |  Yes |
| Default aggfunc |  No | mean |

---

# Default Parameters (Important)

pivot():

df.pivot(
    index=None,
    columns=None,
    values=None
)

No defaults are useful → you MUST specify all 3

---

pivot_table():

df.pivot_table(
    index=None,
    columns=None,
    values=None,
    aggfunc="mean"
)

Default aggregation = mean

# SESSION 12 — GroupBy & Aggregations in Pandas

## Why GroupBy is Important

In real-world data analysis, we often need to answer questions like:

• Total revenue per region  
• Average salary per department  
• Maximum sales per product  

This requires:

-  Grouping data  
-  Applying calculations  

---

# What is GroupBy?

GroupBy means:

-  Split data into groups  
-  Apply a function  
-  Combine results  

---

# Real-Life Analogy

Think of:

Class of students → grouped by section

Then:

• Calculate average marks per section  
• Find top student per section  

---

# One-Line Definition

GroupBy = Split + Apply + Combine

# Basic Syntax

df.groupby("column")

---

# Important

groupby() does NOT compute anything immediately

It creates a grouped object.

You must apply functions like:

• sum()  
• mean()  
• count()  
• max()  
• min()

```python
import pandas as pd

data = {
    "Region": ["North","South","North","South","East"],
    "Category": ["A","A","B","B","A"],
    "Revenue": [100,200,150,250,300]
}

df = pd.DataFrame(data)

df
```

# Example 1 — Group by One Column

Group data by Region

```python
df.groupby("Region").sum()
```

# Parameters of groupby()

df.groupby(by, axis=0, level=None, as_index=True, sort=True, group_keys=True, observed=False, dropna=True)

---

# Parameter Explanation

1. by  
Column name(s) used for grouping  

Example:
"groupby('Region')"

---

2. axis (default = 0)  
0 → group rows  
1 → group columns  

---

3. level  
Used with MultiIndex (advanced)

---

4. as_index (default = True)

True → group column becomes index  
False → group column remains normal column  

---

5. sort (default = True)

True → sort group keys  
False → keep original order  

---

6. group_keys (default = True)

Controls whether group labels appear in output

---

7. observed (default = False)

Used with categorical data

---

8. dropna (default = True)

Drop missing group keys

```python
# as_index=False example

df.groupby("Region", as_index=False).sum()
```

# Example 2 — Group by Multiple Columns

Group by Region and Category

```python
df.groupby(["Region","Category"]).sum()
```

# Aggregation Functions

Common functions:

• sum()  
• mean()  
• count()  
• max()  
• min()  
• std()

---

# Example

```python
df.groupby("Region")["Revenue"].mean()
```

# agg() Function

Allows multiple aggregations

---

# Syntax

df.groupby("column").agg(function)

---

# Example

```python
df.groupby("Region").agg({
    "Revenue": "sum"
})
```

# Multiple Aggregations

We can apply multiple functions

```python
df.groupby("Region").agg({
    "Revenue": ["sum", "mean", "max"]
})
```

# Rename Aggregation Output

Using named aggregations

```python
df.groupby("Region").agg(
    Total_Revenue=("Revenue","sum"),
    Avg_Revenue=("Revenue","mean")
)
```

# Revenue by Region & Category

Goal:

Find total revenue grouped by Region and Category

```python
df.groupby(["Region","Category"])["Revenue"].sum()
```

# Reset Index (Important Step)

After grouping, index becomes hierarchical

Use:

reset_index()

```python
# result = df.groupby(["Region","Category"])["Revenue"].sum()
result = df.groupby(["Region","Category"])["Revenue"].sum().reset_index()

result
```

# Real-World Use Cases

• Sales analysis by region  
• Customer segmentation  
• Product performance  
• Financial reporting  

---

# Common Mistakes

• Forgetting aggregation after groupby  
• Confusion with multi-index  
• Not using reset_index()  

# Summary

GroupBy → grouping data  

Aggregation → applying functions  

agg() → multiple calculations  

Multi-group → more granular analysis  

---

# Key Insight

GroupBy is the backbone of data analytics in Pandas
