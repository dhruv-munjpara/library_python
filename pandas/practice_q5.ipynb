# Question 10 — Full Workflow (Easy but Long)

Create:

customers:

| ID | Name |
|----|------|
| 1 | Aman |
| 2 | Priya |
| 3 | Rahul |

sales:

| ID | Product | Amount |
|----|--------|--------|
| 1 | A | 100 |
| 1 | B | 200 |
| 2 | A | 300 |

Tasks:

1 Merge customers and sales  
2 Replace missing Amount with 0  
3 Create pivot table:  
   • index = Name  
   • columns = Product  
   • values = Amount  
4 Fill missing pivot values with 0  

```python
import pandas as pd
customers=pd.DataFrame({"customer_id":[1,2,3],
                        "name":["aman","priya","rahul"]})
orders=pd.DataFrame({"customer_id":[1,2,4],
                        "amount":[500,700,300]})
```

```python
# task 1.1
pd.merge(customers,orders,how="inner",on="customer_id")
```

```python
# task 1.2
pd.merge(customers,orders,how="left",on="customer_id")
```

```python
# task 1.3
pd.merge(customers,orders,how="outer",on="customer_id")
```

```python
# q2-merge and analyze
# 2.1
pd.merge(customers,orders,how="left",on="customer_id")
```

```python
# 2.2
print("customers with missing order values")
merged=pd.merge(customers,orders,how="left",on="customer_id")
missing_orders=merged[merged["amount"].isna()]
missing_orders
```

```python
# 2.3
print("replce missing values with 0")
merged["amount"]=merged["amount"].fillna(0)
merged
```

```python
print("calculate total order amount")
total_amount=merged.groupby("name")["amount"].sum()
total_amount
```

```python
# q10 full workflow
customers1=pd.DataFrame({"id":[1,2,3],
                         "name":["aman","priya","rahul"]})

sales1=pd.DataFrame({"id":[1,1,2],
                    "product":["A","B","A"],
                    "amount":[100,200,300]})

# task 10.1
print("merge customers and sales:")
merged=pd.merge(customers1,sales1,on="id",how="left")
merged
```

```python
# task 10.2
print("replace missing amount with 0:")
merged["amount"]=merged["amount"].fillna(0)
merged
```

```python
# task 10.3,10.4
print("create a pivot table:")
pivot=pd.pivot_table(merged,index="name",columns="product",values="amount",aggfunc="sum",fill_value=0)
pivot
```

```python
# pandas practice q set-groupby & aggregations 
# q1
df13=pd.DataFrame({"region":["north","south","north","east"],
                   "revenue":[100,200,150,300]})

print("group by region")
df13.groupby("region")["revenue"].sum()
```
