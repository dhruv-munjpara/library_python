>Handling miissing  values in categorical columns:
pandas supports munltiple data types in a dataframe sucu as:
    #numerical:Age,salary
    #categorical:city,gender,product
    #boolean: True/False
    #Datetime:Dates
missing values (NaN) can exist in any type of column,not just numerical
______________________________________________
important clarification:
 #dropna() works on all columns
 #fillna() works on all columns

 BUT:
 the way we fill missing values depends on the data type
 _______________________________________________
 handling numerical vs categorical data
  numerical columns:
    #replace with mean
    #replace with median
 categorical columns:
    #replace with mode(most frequent value)
    #repllace with constant like "unknow"
________________________________________________
why mean dose not work for categorical data

example:
city column:
delhi,mumbai,NaN
we cannot calculate mean of "delhi" ad "mumbai"

so we use:
mode(most frequant value)
constant value like "unknow"


```python
import pandas as pd
import numpy as np

data3= {
    "Customer": ["Aman", "Priya", "Rahul", "Sneha"],
    "Age": [25, np.nan, 30, np.nan],          # Numerical
    "City": ["Delhi", np.nan, "Delhi", "Mumbai"]  # Categorical
}

df3 = pd.DataFrame(data3)

print("Original DataFrame:\n")
print(df3)
```

1.filling numerical coilumn(Age):
use mean or median

```python
df3["Age"]=df3["Age"].fillna(df3["Age"].mean())
```

2.filling categorical column using mode.
mode=most frequent value

```python
df3["City"]=df3["City"].fillna(df3["City"].mode())
df3
```

3.filling categorical column using constant.
insted of mode,we can use a constant value.
example:"unknow"

```python
df3["City"]=df3["City"].fillna("unknow")
df3
```

4.using fillna() for multuple columns.

we can handel both numeroical and categorical column todether.

example:
#age-replace with mean
#city-replace with "unknow"

```python
df3=df3.fillna({"Age":df3["Age"].mean(),
                "City":"unknow"})
df3
```

>what is outliers??
outliers are extreme values that differ significantly from other observations.

examples:[10,12,11,13,500]
so value 500 is an outlier.

outliers can occure due to:
    #measurement errors
    #data entry errors
    #unusual events

>detecting outliers using iqr method
iqr=interquartile range
formula:iqr=q3-q1
lower bound=q1-1.5*iqr
upper bound=q3+1.5*iqr

```python
salse=pd.Series([100,120,130,110,115,500])

q1=salse.quantile(0.25)
q3=salse.quantile(0.75)

iqr=q3-q1

lower=q1-1.5*iqr
upper=q3+1.5*iqr

outliers=salse[(salse<lower)|(salse>upper)]
outliers
```

>Z-score concept:
Z-score mesures how far a value is from the mean

formula:Z=(X-Mean)/standard deciation

values with |Z|>3 are often considered outliers.

```python
mean=salse.mean()
std=salse.std()

z_score=(salse-mean)/std
z_score
```

session 10  ==merging & joining dataframe

why merging is imporatant
    in real-world scenarios,data is often stored in multiple tables/files.

    Examples:

Customer Table:

| Customer_ID | Name |
|-------------|------|
| 1 | Aman |
| 2 | Priya |

Sales Table:

| Customer_ID | Amount |
|-------------|--------|
| 1 | 500 |
| 2 | 700 |
    
to analyze data properly,we need to combine these tables.
__________________________________________
solution:
pandas provides: merge()
____________________________________________
key concept:
merge is similar to sql join

merge () function:
syntax:pd.merge(left_df,right_df,how="type",on="column")
___________________________________________
parameters:
left df=first dataframe
right df=second dataframe

how=type of join
    #inner=common values only
    #left=all values from left
    #right=all values from right
    #outer=all values from both

on=column name used for joining
____________________________________________
use case:
>combine customer data with sales
>combine dataset from munltiple sources


```python
customers=pd.DataFrame({
    "customer_ID":[1,2,3],
    "name":["aman","priya","rahul"]
})

salse=pd.DataFrame({
    "customer_ID":[1,2,4],
    "amount":[500,700,300]
})

customers,salse
```

```python
# 1.inner join :keep only matching values in both tables:

pd.merge(customers,salse,how="inner",on="customer_ID")
```

```python
# 2.left join:keep all  rows from left table
pd.merge(customers,salse,how="left",on="customer_ID")
```

```python
# 3 right join:keep all rows from right table
pd.merge(customers,salse,how="right",on="customer_ID")
```

```python
# outer join:keeps all rows from both tables.
pd.merge(customers,salse,how="outer",on="customer_ID")
```

merge on multiple columns:
syntax:pd.merge(df1,df2,on=["col1","col2"])
________________________________________________
use case:
when one column is not enough to uniquely indentify data.

```python
df10=pd.DataFrame({
    "id":[1,1,2],
    "city":["delhi","mumbai","delhi"],
    "sales":[100,200,300]
})

df11=pd.DataFrame({
    "id":[1,2],
    "city":["delhi","delhi"],
    "target":[150,350]
})
pd.merge(df10,df11,on=["id","city"])
```

default behavior of pd.merge()
when you do:
pd.merge(df1,df2,on["id","city"])
and do not specify how
___________________________________________
defautlt join type
the default join is:
inner join
_________________________________________
what is inner join?
it keeps only those rows where:
#values match in both dataframes
#based on the given key(s)
________________________________________
important rule
if no match exists=row is dropped
________________________________________
syntax reminder
pd.merge(df1,df2,how="inner",on=["id","city"])
is same as
pd.merge(df1,df2,on=["id","city"])
____________________________________________
demo-customer+sales merge
goal:
combine customer details with sales data.

```python
merged_df=pd.merge(customers,salse,how="left",on="customer_ID")
merged_df
```

# SESSION 12 — GroupBy & Aggregations in Pandas

## Why GroupBy is Important

In real-world data analysis, we often need to answer questions like:

• Total revenue per region  
• Average salary per department  
• Maximum sales per product  

This requires:

-  Grouping data  
-  Applying calculations  

---

# What is GroupBy?

GroupBy means:

-  Split data into groups  
-  Apply a function  
-  Combine results  

---

# Real-Life Analogy

Think of:

Class of students → grouped by section

Then:

• Calculate average marks per section  
• Find top student per section  

---

# One-Line Definition

GroupBy = Split + Apply + Combine


# Basic Syntax

df.groupby("column")


# Important

groupby() does NOT compute anything immediately

It creates a grouped object.

You must apply functions like:

• sum()  
• mean()  
• count()  
• max()  
• min()


```python
import pandas as pd
df39=pd.DataFrame({"region":["north","south","north","south","east"],
                   "category":["A","A","B","B","A"],
                   "revenue":[100,200,150,250,300]})
df39
```

```python
# ex 1 -group by one cloumn
 
# group data by region
df39.groupby("region").sum()
```

# Parameters of groupby()

df.groupby(by, axis=0, level=None, as_index=True, sort=True, group_keys=True, observed=False, dropna=True)

---

# Parameter Explanation

1. by  
Column name(s) used for grouping  

Example:
"groupby('Region')"

---

2. axis (default = 0)  
0 → group rows  
1 → group columns  

---

3. level  
Used with MultiIndex (advanced)

---

4. as_index (default = True)

True → group column becomes index  
False → group column remains normal column  

---

5. sort (default = True)

True → sort group keys  
False → keep original order  

---

6. group_keys (default = True)

Controls whether group labels appear in output

---

7. observed (default = False)

Used with categorical data

---

8. dropna (default = True)

Drop missing group keys

```python
# as_index=false example

df39.groupby("region",as_index=False).sum()
```

```python
# ex 2 ==group by multiple column
# group by region and category
df39.groupby(["region","category"]).sum
```

# Aggregation Functions

Common functions:

• sum()  
• mean()  
• count()  
• max()  
• min()  
• std()

---

# Example

```python
df39.groupby("region")["revenue"].mean()
```

# agg() Function

Allows multiple aggregations

---

# Syntax

df.groupby("column").agg(function)

---

# Example

```python
df39.groupby("region").agg({"revenue":"sum"})
```

```python
# multiple aggregation
# we can apply multiple function
df39.groupby("region").agg({"revenue":["sum","mean","max"]})
```

```python
# rename aggregation output
# using named aggregations
df39.groupby("region").agg(total_revenue=("revenue","sum"),
                           avg_revenue=("revenue","mean"))
```

```python
# revenue by region & category
# goal:find total revenue grouped by region and category
df39.groupby(["region","category"])["revenue"].sum()
```

```python
# reset index (important step)
# after grouping,index becomes hierarchical
# use: reset_index()
# result = df.groupby(["Region","Category"])["Revenue"].sum()
result = df39.groupby(["region","category"])["revenue"].sum().reset_index()

result
```

# Real-World Use Cases

• Sales analysis by region  
• Customer segmentation  
• Product performance  
• Financial reporting  

---

# Common Mistakes

• Forgetting aggregation after groupby  
• Confusion with multi-index  
• Not using reset_index()  


# Summary

GroupBy → grouping data  

Aggregation → applying functions  

agg() → multiple calculations  

Multi-group → more granular analysis  

---

GroupBy is the backbone of data analytics in Pandas
