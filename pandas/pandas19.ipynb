```python
merged=pd.merge(customers,orders,how="outer",on="Customer_ID")
merged
```

```python
merged.drop_duplicates()
```

```python
clean_merged=merged.replace("electronics","Electronics")
clean_merged
```

```python
clean_merged1=clean_merged.replace("ELECTRONICS","Electronics")
clean_merged1
```

```python
clean_merged1["Amount"]=clean_merged1["Amount"].fillna(clean_merged1["Amount"].mean())
clean_merged1
```

```python
clean_merged1["Region"]=clean_merged1["Region"].fillna("Unknow")
clean_merged1
```

```python
clean_merged1["Order_Date"]=pd.to_datetime(clean_merged1["Order_Date"])
clean_merged1
```

```python
clean_merged1["Amount_Category"] = np.where(clean_merged1["Amount"] > 50000, "High", "Low")
clean_merged1
```

```python
clean_merged1["Tax"] =clean_merged1["Amount"] * 0.18
clean_merged1["Total_Amount"] = clean_merged1["Amount"] + clean_merged1["Tax"]
clean_merged1
```

```python
clean_merged1["Order_Date"] = pd.to_datetime(clean_merged1["Order_Date"])
clean_merged1
```

```python
clean_merged1["Month"] = clean_merged1["Order_Date"].dt.month
clean_merged1
```

```python
clean_merged1["year"] = clean_merged1["Order_Date"].dt.year
clean_merged1
```

```python
clean_merged1["day"] = clean_merged1["Order_Date"].dt.day
clean_merged1
```

```python
# total sales per customers
clean_merged1.groupby("Name")["Amount"].sum()
```

```python
# sales per rigion
clean_merged1.groupby("Region")["Amount"].sum()
```

```python
# select per product
clean_merged1.groupby("Product")["Amount"].sum()
```

```python
# region+product
clean_merged1.groupby(["Region","Product"])["Amount"].sum().reset_index()
```

```python
# pivot  table
pivot1=pd.pivot_table(clean_merged1,values="Amount",index="Region",columns="Product",aggfunc="sum",fill_value=0)
pivot1
```

```python
# pivot  table
pivot2=pd.pivot_table(clean_merged1,values="Amount",index="Region",columns="Product",aggfunc="sum",fill_value=0)
pivot2
```

```python
pivot2 = pd.pivot_table(
    clean_merged1,
    values="Amount",
    index="Month",
    columns="Product",
    aggfunc="sum",
    fill_value=0
)

pivot2
```
