```python
import pandas as pd

customers=pd.DataFrame({"customer_id":[1,2,3,4],
                        "region":["north","south","east","west"]})

sales=pd.DataFrame({"customer_id":[1,1,2,3,4],
                    "product":["A","B","A","B","A"],
                    "amount":[100,200,300,150,250]})

customers,sales
```

```python
# value null or not
customers.isnull().sum()
```

```python
# salse data null or not
sales.isnull().sum()
```

```python
# merge dual table
merged=pd.merge(customers,sales,how="outer",on="customer_id")
merged
```
