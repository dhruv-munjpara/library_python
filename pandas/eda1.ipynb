```python
plt.hist(df["SepalLengthCm"],bins=10)
plt.title("histogram")
plt.xlabel("sepal length")
plt.ylabel("frequancy")
plt.show()
# Used for distribution.

plt.boxplot(df["SepalLengthCm"])
plt.title("Box Plot")
plt.show()
# Used for outliers.
```

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=df)
plt.show()
# Find median
# Find quartiles
# Detect outliers
# Understand spread of data

sns.violinplot(x="Species", y="SepalLengthCm", data=df)
plt.show()
# Distribution shape
# Density of data
# Median
# Spread
```

```python
groupby_with_species=df.groupby("Species").agg(totalSepalLengthCm=("SepalLengthCm","sum"),
                                               totalSepalWidthCm=("SepalWidthCm","sum"),
                                               totalPetalLengthCm=("PetalLengthCm","sum"),
                                               totalPetalWidthCm=("PetalWidthCm","sum"))
groupby_with_species
# Data summary
# Comparing groups
# Reporting totals
# Feature analysis
# EDA summary tables
```
