session 6==handling duplicate and outliers::
this is the common problem in dataset
    1.duplicate record
    2.outliers

>handling duplicate rows:
function:drop_duplicates()
syntax:df.drop_duplicates()

parameter::
specify columns to check duplicates

keep:
"first"=keeo first occurrence
"last"=keeo last occurrence
"false"=remove all duplicates



>detecting outliers using iqr method:

iqr=interquartile range
formula:iqr=q3-q1
lowew bound=q1-1.5*iqr
upper bound=q3+1.5*iqr

outliers=column[(column<lower)|(column>upper)]



>z-score concept

z-score mesuare how far a values is from the mean.

formula:z=(x-mean)/standard deviation
values with|z|>3 are often consider outliers.

session 10==merging & joining datframe:

>merge() function:

syntax:
pd.merge(left_df,right_df,how="type",on="column")

parameter:
left_df=first dataframe
right_df=second dataframe

how=type of join:
1.inner=common values
2.left=all values from left
3.right=all values from right
4.outer=all values from both

on=column name used for joining

inner_join:
pd.merge(customer,sales,how="inner",on="customer_id")
left:
pd.merge(customer,sales,how="left",on="customer_id")
right:
pd.merge(customer,sales,how="right",on="customer_id")
outer:
pd.merge(customer,sales,how="outer",on="customer_id")

>merge on multiple column:
syntax:
pd.merge(df1,df2,on=["col1","col2"])

ex:
pd.merge(df1,df2,on=["id","city"])

>default behavior of pd.merge()
default its take inner join

session 12-groupby & aggregation in pandas:

>why groupby in i,portant
    in real-world data analysis,we often need to answer question like:
    >total revenue per region
    >average salary per region
    >maximum sales per product
this requires:
    >GROUPing data
    >applying calculation

what is groupby:
>split data in groups
>apply a function
>combine results  

>one-line defination:
groupby=split+apply+combine

>basic syntax
df.groupby("column")

important:
you must apply function:
1.sum()
2.mean()
3.count()
4.max()
5.min()

>ex:1-group by one column:
group data by region:
ex:df.group("region").sum()

parameter of groupby():
EX:df.groupby(by,axis=0,level=none,as_index=True,sort=True,group_keys=True,observed=Fales,droupna=True)

parameter explanation:
1.by:column name(s) used for grouping
    ex:"groupby("region")"

2.axis(default=0)
    0=group rows
    1=group column

3.level:used with multiindex(advanced)

4.as_index(default=True)
true->group column becomes index
False->group column remains normal column

5.sort(default=True)
True->sort group keys
false->keep original order

6.group_keys(default=true):
controls whether group labels apper in output

7.observed(default=false)
used with categories data

8.dropna(default=true)
drop missing group keys

ex:df.group("region",as_index=false).sum()
ex:df.group(["region","category"]).sum()

aggregation function:
1.sum()
2.mean()
3.count()
4.max()
5.min()
6.std()

ex:df.groupby("region")["revenue"].mean()

>agg() function
allows multiple aggregations

syntax:df.groupby("column").agg(function)

ex:df.groupby("region").agg({"revenue":"sum"})

>multiple aggregations:we can apply multiple function

ex:
df.groupby("region").agg({"revenue":["sum","mean","max"]})

>rename aggregation output:
using named aggregations

ex:
df.groupby("region").agg(
    total_revenue=("revenue","sum"),
    avg_revenue=("revenue","mean")
)

>revenue by region & category:
find total revenue grouped by region and category

ex:
df.groupby(["region","category"])["revenue"].sum()



>reset index(important step):
after groping index becomes hierachical

use:reset_index()

ex:
df.groupby(["REGION","category"])["revenue"].sum().reset_index()
