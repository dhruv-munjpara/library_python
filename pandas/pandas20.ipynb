```python
import pandas as pd
import numpy as np

# Customer dataset
customers = pd.DataFrame({
    "Customer_ID": [1,2,3,4,5,6],
    "Name": ["Aman Sharma","Priya Patel","Rahul Verma","Sneha Shah","Karan Mehta","Riya Jain"],
    "Region": ["North","South","East","West","North",None]
})

# Orders dataset
orders = pd.DataFrame({
    "Customer_ID": [1,1,2,3,4,4,5,6,6],
    "Product": ["Laptop","Mouse","Laptop","Tablet","Laptop","Tablet","Mouse","Laptop","Mouse"],
    "Category": ["Electronics","electronics","Electronics","electronics","ELECTRONICS","electronics","electronics","Electronics","electronics"],
    "Amount": [50000,2000,55000,30000,60000,25000,1500,70000,np.nan],
    "Order_Date": ["2024-01-01","2024-01-05","2024-02-01","2024-02-10","2024-03-01","2024-03-05","2024-03-10","2024-04-01","2024-04-05"]
})

customers, orders
```

```python
# customer data clean
customers
```

```python
customers["Region"]=customers["Region"].fillna("Unknow")
customers
```

```python
# order data clean
orders
```

```python
orders["Amount"]=orders["Amount"].fillna(orders["Amount"].mean())
orders
```

```python
orders["Category"]=orders["Category"].replace("ELECTRONICS","Electronics")
orders
```

```python
orders["Category"]=orders["Category"].replace("electronics","Electronics")
orders
```

```python
orders["Order_Date"]=pd.to_datetime(orders["Order_Date"])
orders 
```

```python
merged_data=pd.merge(customers,orders,how="outer",on="Customer_ID")
merged_data
```

```python
merged_data.groupby(["Name", "Region"]).agg({
    "Category": "first",
    "Amount": "sum"
}).reset_index()
```

```python
pd.pivot_table(merged_data,
               values="Amount",
               index="Region",
               columns="Product",
               aggfunc="sum",
               fill_value=0,
               margins=True)
```
